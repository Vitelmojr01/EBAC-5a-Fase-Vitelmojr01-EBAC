````python
import pandas as pd
import numpy as np

# Gerando acertos aleatórios por assunto para 100 alunos, respeitando o número de questões por assunto
np.random.seed(42)
df = pd.DataFrame({
    'Matemática': np.random.randint(0, 25, 100),
    'Português': np.random.randint(0, 19, 100),
    'Geografia': np.random.randint(0, 9, 100),
    'Inglês': np.random.randint(0, 9, 100),
    'História': np.random.randint(0, 9, 100),
    'Física': np.random.randint(0, 13, 100),
    'Química': np.random.randint(0, 13, 100)
})


# 1. Percentual de acerto por assunto
numero_questoes = {
    'Matemática': 24,
    'Português': 18,
    'Geografia': 8,
    'Inglês': 8,
    'História': 8,
    'Física': 12,
    'Química': 12
}

df_percentual_por_assunto = df.copy()
for assunto in df.columns:
    df_percentual_por_assunto[assunto] = (df[assunto] / numero_questoes[assunto]) * 100



# 2. Total de acertos por aluno
df['Total_Acertos'] = df.sum(axis=1)

# 3. Percentual geral por aluno
total_questoes = sum(numero_questoes.values())  # 90
df['Percentual_Geral'] = (df['Total_Acertos'] / total_questoes) * 100

# 4. Quantos alunos tiveram nota maior que 45?
alunos_aprovados = df[df['Total_Acertos'] > 45].shape[0]


````

````python
import pandas as pd

# 1. Criar DataFrame só com vacinas em menores de 18 anos
df_menores = df[df["paciente_idade"] < 18].copy()
print(df_menores)
````

item 02

````python
# 2. Ver colunas disponíveis
colunas_disponiveis = df_menores.columns
print(colunas_disponiveis)
````

````python
# 3. Contagem da coluna vacina_categoria_nome (não obrigatória para o final, mas útil para análise)
print(df_menores['vacina_categoria_nome'].value_counts())
````

````python
# 4. Criar lista com colunas que identificam os estabelecimentos
colunas_estabelecimento = [
    'estabelecimento_valor',
    'estabelecimento_razaosocial',
    'estalecimento_nofantasia',
    'estabelecimento_municipio_codigo',
    'estabelecimento_municipio_nome',
    'estabelecimento_uf'
]
````

````python
# 5. Selecionar só essas colunas
df_estabelecimentos_menores = df_menores[colunas_estabelecimento]
print(df_estabelecimentos_menores)
````

````python
# 6. Remover duplicatas para ter um único registro por estabelecimento
df_estabelecimentos_menores = df_estabelecimentos_menores.drop_duplicates()
````

````python
# 7. Agrupar e contar quantas vacinas cada estabelecimento aplicou em menores
df_contagem = df_menores.groupby('estabelecimento_valor').size().reset_index(name='qtd_vacinas_menores')
````

````python
# 8. Juntar os dois DataFrames
df_resultado = pd.merge(df_estabelecimentos_menores, df_contagem, on='estabelecimento_valor')
````

````python
# Exibir resultado final
print(df_resultado.head())
print(f"\nTotal de estabelecimentos que vacinaram menores: {df_resultado.shape[0]}")
````
